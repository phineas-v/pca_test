```python
# Clean data
# Check for infinite values
print("Infinite values:")
for col in df.select_dtypes(include=[np.number]).columns:
    inf_count = np.isinf(df[col]).sum()
    if inf_count > 0:
        print(f"{col}: {inf_count}")
# Check for NaN values
print("NaN values:")
for col in df.select_dtypes(include=[np.number]).columns:
    nan_count = df[col].isna().sum()
    if nan_count > 0:
        print(f"{col}: {nan_count}")
```

```python
# Drop columns "Reaction_Time"
df_dropped = df.drop(columns=["Reaction_Time"])
```

```python
# Min - max Normalization
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler    
```

```python
scaler = StandardScaler() # Refer from paper for normalization technique
df_scaled = scaler.fit_transform(df_dropped.select_dtypes(include=[np.number])) 
df_scaled = pd.DataFrame(df_scaled, columns=df_dropped.select_dtypes(include=[np.number]).columns)
# Now df_scaled contains the normalized numerical features
# If needed, concatenate with non-numerical columns
df_final = pd.concat([df_scaled, df_dropped.select_dtypes(exclude=[np.number]).reset_index(drop=True)], axis=1)
```

```python
df_final.head()
```

```python
# Show biplot of features with 2 principal components
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(df_final.select_dtypes(include=[np.number]))
# Loadings
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
# Biplot
plt.figure(figsize=(10, 7))
plt.scatter(loadings[:, 0], loadings[:, 1], alpha=1)
column_names = ["OBD_Engine_Load", "OBD_Speed_Km", "OBD_Engine_Coolant_Temp_C","OBD_Engine_RPM", "Device_Trip_Dist_Km", "OBD_Fuel_Flow_CCmin", "OBD_Air_Pedal", "Acceleration_kmhs"]
for i, feature in enumerate(column_names):   
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='r', alpha=0.5)
    plt.text(loadings[i, 0], loadings[i, 1], feature, color='g', ha='center', va='center')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Biplot')
plt.grid()
plt.show()
```

```python
# Combined variance score of first 2 principal components
explained_variance = pca.explained_variance_ratio_
print("Explained variance by each component:", explained_variance)
print("Combined explained variance of first 2 components:", np.sum(explained_variance[:2]))
```

```python
# Display loadings of each feature for first 2 principal components
loading_df = pd.DataFrame(loadings, index=df_final.select_dtypes(include=[np.number]).columns, columns=['PC1', 'PC2'])

# Add a column that find the vector magnitude of the first 2 principal components
loading_df['Magnitude'] = np.sqrt(loading_df['PC1']**2 + loading_df['PC2']**2)

# Sort by magnitude
loading_df = loading_df.sort_values(by='Magnitude', ascending=False)
print("Sorted Loadings by Magnitude:")
print(loading_df)
```

```python
y = pd.Series(df['OBD_CO2_gkm_Instant'], name="CO2_Instant")
X = df_final.select_dtypes(include=[np.number]).drop(columns=['OBD_CO2_gkm_Instant'])

```

```python
# Feature Importance using LASSO
from sklearn.linear_model import LassoCV

lasso = LassoCV(cv=5, random_state=0)
lasso.fit(X, y)

coef = pd.Series(lasso.coef_, index = X.columns)
selected_features = coef[coef != 0].index.tolist()

print("Selected features:", selected_features)
print("\nLASSO coefficients:")
print(coef.sort_values(ascending=False))
print(f"\nOptimal alpha (): {lasso.alpha_:.4f}")

```

```python
# Random Forest Regressor for Feature Importance
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=200,
    random_state=42,
    n_jobs=-1)
model.fit(X, y)
```

```python
importances = pd.Series(model.feature_importances_, index=X.columns)
importances = importances.sort_values(ascending=False)

print("Feature Importances:\n", importances)

# Optionally select top N features
top_features = importances.head(5).index.tolist()
print("\nTop 5 important features:", top_features)
```
